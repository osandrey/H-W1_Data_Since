```python
#Скільки книг з рейтингом 4.9 потрапило до рейтингу у 2010 та 2011 роках (використовуйте | або функцію isin)? Відповідь:

rating = data[(data['user_rating'] == 4.9) & (data['year'].isin([2010, 2011]))]

print(f'Скільки книг з рейтингом 4.9 потрапило до рейтингу у 2010 та 2011 роках ? Відповідь: {len(rating)}')
```

```python
#І насамкінець, давайте відсортуємо за зростанням ціни всі книги, які потрапили до рейтингу в 2015 році і коштують дешевше за 8 доларів (використовуйте функцію sort_values).

filtered_books = data[(data['year'] == 2015) & (data['price'] < 8)]
answer = filtered_books.sort_values(by='price', ascending=True)
print(f'Яка книга остання у відсортованому списку? Відповідь: {answer.tail(1)}')
```

```python
#Для початку давайте подивимося на максимальну та мінімальну ціни для кожного з жанрів (використовуйте функції groupby та agg, для підрахунку мінімальних та максимальних значень використовуйте max та min). Не беріть усі стовпці, виберіть тільки потрібні вам

genre_price = data.groupby('genre')[['price']].agg(['min', 'max'])
print(f'Подивимося на максимальну та мінімальну ціни для кожного з жанрів: {genre_price}')
print(f'Відповідь: Максимальна ціна для жанру Fiction: {genre_price.loc["Fiction", ("price", "max")]}')
print(f'Відповідь: Vінімальна ціна для жанру Fiction: {genre_price.loc["Fiction", ("price", "min")]}')
print(f'Відповідь: Максимальна ціна для жанру NoN Fiction: {genre_price.loc["Non Fiction", ("price", "max")]}')
print(f'Відповідь: Vінімальна ціна для жанру NoN Fiction: {genre_price.loc["Non Fiction", ("price", "min")]}')
```

```python
#Тепер створіть новий датафрейм, який вміщатиме кількість книг для кожного з авторів (використовуйте функції groupby та agg, для підрахунку кількості використовуйте count). Не беріть усі стовпці, виберете тільки потрібні

author_books_count = data.groupby('author')['name'].agg('count').reset_index()
author_books_count.rename(columns={'name': 'book_count'}, inplace=True)
print(author_books_count)
print(f'Розмірність таблиці: {author_books_count.shape}')
most_prolific_author = author_books_count['book_count'].idxmax()
author = author_books_count.loc[most_prolific_author, 'author']
books = author_books_count.loc[most_prolific_author, 'book_count']

print(f'Автор з найбільшою кількістю книг: {author} \nКількість книг: {books}')
```

```python
#Тепер створіть другий датафрейм, який буде вміщати середній рейтинг для кожного автора (використовуйте функції groupby та agg, для підрахунку середнього значення використовуйте mean). Не беріть усі стовпці, виберете тільки потрібні

avg_rating = data.groupby('author')['user_rating'].agg('mean').reset_index()
avg_rating.columns = ['author', 'average_rating']
print(avg_rating)
min_avg_rating_author = avg_rating[avg_rating['average_rating'] == avg_rating['average_rating'].min()]
auth = min_avg_rating_author['author'].iloc[0]
rating = min_avg_rating_author['average_rating'].iloc[0]
print(f'У якого автора середній рейтинг мінімальний? Відповідь: {auth}')
print(f'Який у цього автора середній рейтинг? Відповідь: {rating}')


```

```python
#З'єднайте останні два датафрейми так, щоб для кожного автора було видно кількість книг та середній рейтинг (Використовуйте функцію concat з параметром axis=1). Збережіть результат у змінну

concated_df = pd.concat([author_books_count.set_index('author'), avg_rating.set_index('author')], axis=1)
print(concated_df)
```

```python
#Відсортуйте датафрейм за зростаючою кількістю книг та зростаючим рейтингом (використовуйте функцію sort_values)
sorted_df = concated_df.sort_values(by=['book_count', 'average_rating'])
print(sorted_df)
sorted_df.reset_index(inplace=True)
first_author = sorted_df.iloc[0]['author']
print(f"Який автор перший у списку? Відповідь: {first_author}")
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(10, 6))
plt.plot(sorted_df['book_count'], sorted_df['average_rating'], marker='o', color='purple', label='Кількість книг')
plt.title('Залежність середнього рейтингу від кількості книг')
plt.xlabel('Кількість книг')
plt.ylabel('Середній рейтинг')
plt.legend()
plt.grid(True)
plt.show()

# Виведемо ім'я першого автора у списку
first_author = sorted_df.iloc[0]['author']
print(f"Який автор перший у списку? Відповідь: {first_author}")



```
