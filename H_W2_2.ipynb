```python
#Видаліть усі рядки у вихідній таблиці за допомогою методу dropna
cleaned_df = df.dropna()
# print(cleaned_df)
#Визначте новий розмір таблиці за допомогою методу shape
print(cleaned_df.shape)
```

```python
#Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python
python_data = cleaned_df[cleaned_df['Язык.программирования'] == 'Python']
# print(python_data)
print(python_data['Язык.программирования'])
```

```python
#Визначте розмір таблиці python_data за допомогою методу shape

print(python_data.shape)
```

```python
#Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"/ Должность
grouped_by_position = python_data.groupby('Должность')
for position, group_data in grouped_by_position:
    print(f"Position: {position}")
    print(group_data)
```

```python
#Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"

aggregated_data = grouped_by_position.agg({'Зарплата.в.месяц': ['min', 'max']})

print(aggregated_data)
```

```python
aggregated_df = pd.DataFrame(aggregated_data)

aggregated_df.plot(kind='bar', figsize=(10, 6))
plt.title('Min-Max Salary by Job Position')
plt.xlabel('Job Position')
plt.ylabel('Salary')
plt.show()



```

```python
#Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"

def fill_avg_salary(group):
    avg_salary = group['Зарплата.в.месяц'].mean()
    group['avg'] = avg_salary
    return group


python_data = grouped_by_position.apply(fill_avg_salary)
print(python_data)
```

```python
#Створіть описову статистику за допомогою методу describe для нового стовпчика.
avg_column_description = python_data['avg'].describe()
print(avg_column_description)
```
